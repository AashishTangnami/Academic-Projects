# Read Data

Very first step is to import raw data into the code base and there are multiple ways of doing so.

```python
with open('emerson_self-reliance.txt') as f:
    tmp_text = f.read()
```

```python
tmp_text
```

```python

"""
Purpose of using inflect library is to convert the numbers in the text dataset into words.

We are trying to convert the numbers into words 

For example:  
text input : Abhiyan has 0 marks in dataware house assignment.
-> using inflect library function : number_to_words()
text ouptut: Abhiyan has no marks in dataware house assignment.

"""
# import the inflect library
import inflect

inflect_engine = inflect.engine()
 
# convert number into words
def convert_number_to_word(text):
    # split string into list of words
    raw_text = text.split()
    # initialise empty list
    new_text_data = []
 
    for word in raw_text:
        # if word is a digit, convert the digit
        # to numbers and append into the new_string list
        if word.isdigit():
            temp_word = inflect_engine.number_to_words(word)
            new_text_data.append(temp_word)
 
        # append the word as it is
        else:
            new_text_data.append(word)
 
    # join the words of new_string to form a string
    raw_text = ' '.join(new_text_data)
    return raw_text
```

```python
new_text = convert_number_to_word(tmp_text)
new_text
```

### Observation:

1. We can notice that numeric value 1841 has been changed into word value.

# Lower Casing

This is foremost important step in the text cleaning so that all the text format for same text is treated as same. 

For eg. 'Aashish', 'AASHISH' and  'aashish' is treated as same. If not then the all three words are represented as three different words in the vector space model that results in more dimensions.

```python
def lower_text(text):
  # custom function to lower the text format.
  # by default most of the tokenizer libraries lowers the case.
    text = text.lower()
    return text

lower_text = lower_text(new_text)
lower_text
```

### Observation:

1. All the text has been changed into lower cases. 

#Remove Punctuations

The process of removing punctuation from the text data is to standardize the text so that we can treat words such as 'Hi!' and 'Hi' as same word.

- "string.punctuation" in python contains the following punctuation symbols - "!"#$%&\'()*+,-./:;<=>?@[\\]^_{|}~`" 

```python
def remove_punctuation(text):
    # function to remove the punctuation from the text.
    trans = str.maketrans('','', string.punctuation)
    return text.translate(trans)

punct_text = remove_punctuation(lower_text)
punct_text
```

### Observation

1. All the special characters or symbols has been removed from the text.

2. Leading and ending spaces has been eliminated. 

3. On third line - has not been acknowledged as punctuation.

#Tokenization

The process of converting or splitting the sentences into words without any blanks or tabs is called tokenization.



```python
# importing the tokenizer from nltk

from nltk.tokenize import word_tokenize
# nltk.download('punkt')

def text_tokenize(text):
    token_text = word_tokenize(text)
    return token_text
```

#Remove Stopwords

Stopwords are the most common words that do not carry important meaning thus are usually removed from texts.

```python
# importing stopwords from nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

def filtered_text(token_text):
    # function to filterout the stopwords from the tokens.
    stop_words = set(stopwords.words("english")) # set of stopwords.
    filtered_text = [token for token in token_text if token not in stop_words]
    return filtered_text
```

```python
nonstop_text_sentence_token = filtered_text(corpus)
nonstop_text_sentence_token
```

```python
nonstop_text_token = filtered_text(token_text)
nonstop_text_token
```

### Observations:

1. All the existing stopwords from the tokenized text data has been removed.

#Stemming

Stemming is a process of transforming or reducing a word to its root form.
for example the words "helping" and "helper" share root word "help".
Stemming enables you to focus on a word's core meaning rather than all the specifics of its usage. We are using stemmers available in NLTK.


Stemming has the drawback of altering the meaning of the root form of the word or preventing it from becoming a genuine english word.


```python
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()
 
# stem words in the list of tokenized words
def stem_words(text):
    stemmed_text = [stemmer.stem(word) for word in text]
    return stemmed_text
```

```python
stemmed_text = stem_words(nonstop_text_token)
stemmed_text
```

### Observation

1. Each word from the corpus are seperated as token.

#Lemmatize

Lemmatizing is similar to stemming in that it reduces words to their essence, but it gives you a full English word that stands alone as opposed to only a part of a word like "discoveri."

Unlike like stemming, lemmatizing reduces the plural word example: 'helps' and 'carves' becomes singular word as 'help' and 'scarf'



```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def lemmatize_word(text):
    # this function is to lemmatize tokenized words.
    lemmatize_text = [lemmatizer.lemmatize(word) for word in text]
    return lemmatize_text

lemmatized_text = lemmatize_word(nonstop_text_sentence_token)
```

```python
lemmatized_text
```

### Observations

1. Lemmatization is preferred over stemming because lemmatization does a morphological analysis of the words.

2. Lemmatization considers the context and converts the word to its meaningful base form, whereas stemming just removes the last few characters, often leading to incorrect meanings and spelling errors.

#Vectorizing

Word vectorization is a methodology in NLP to map words or phrases from vocabulary to a corresponding vector of real numbers which used to find word predictions word similarities/semantics.

Basically process of converting words into numbers are called Vectorization.

We are using CountVectorizer
1. Simplest ways of doing text vectorization.

2. Creates a document term matrix of dummy variables if particular word appears in the document.

3. Count vectorizer will fit and learn the word vocabulary and try to create a document term matrix in which the individual cells denote the frequency of that word in a particular document, which is also known as term frequency and the columns are dedicated to each word in the corpus.

```python
from sklearn.feature_extraction.text import CountVectorizer
# vectorizing the stemmed_text of the tokenized sentences.
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(stemmed_text)
vectorizer.get_feature_names_out()

print(X.toarray())
```

```python
# vectorizing the lemmatized text of the sentences as token from the dataset.
# this technique is used to find the relevance in meaning while doing text summarization in ML models.
vectorizer2 = CountVectorizer(analyzer='word', ngram_range=(2, 2))
X2 = vectorizer2.fit_transform(lemmatized_text)
vectorizer2.get_feature_names_out()
print(X2.toarray())
```
